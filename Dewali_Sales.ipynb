```python
d.columns
```

```python
state = d.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(15)

# to set chart size
sns.set(rc ={'figure.figsize': (24,4)})
sns.barplot(data = state, x = 'State', y = 'Orders')
 

```

# here we can see in first place is uttarpradesh in highest count of orders have ordered 

```python
state = d.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(15)


# to set chart size
sns.set(rc ={'figure.figsize': (24,4)})
sns.barplot(data = state, x = 'State', y = 'Amount')
 
```

# here we can see uttar pradesh has high purchase power,maharastra,karnataka...

# martial status

```python
m = sns.countplot(data = d, x = "Marital_Status")

sns.set(rc = {"figure.figsize" :(8,5) })
for i in m.containers:
    m.bar_label(i)
```

```python
m = d.groupby(['Marital_Status','Gender'], as_index = False)['Amount'].sum().sort_values(by= "Amount",ascending = False)


sns.set(rc = {"figure.figsize" :(6,5) })
sns.barplot(data = m, x = "Marital_Status", y = "Amount", hue = "Gender")
```

# from the above graph we can say that most of the buyers are married women 

```python
#occuption

sns.set(rc= {"figure.figsize" :(25,7)})
o = sns.countplot(data = d, x = "Occupation")

for i in o.containers:
    o.bar_label(i)
```

# from the above graph we can say that most of the buyers from IT sector  and second highest is Health care domains

```python
am = d.groupby(["Occupation"],as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)

sns.set(rc = { 'figure.figsize' : (25,7)})
sns.barplot(data = am,x= "Occupation", y = "Amount")
```

# first highest money spend to buy products are IT sector sencond Helathcare and third Aviation

# product category

```python
sns.set(rc = {'figure.figsize' : (25,7)})
p = sns.countplot(data = d, x = 'Product_Category')

for i in p.containers:
    p.bar_label(i)
```

# we can say that first most sales done in clothing & apparel 
#second in Food
#third in electronic & Gadgets

```python
product_sales = d.groupby(["Product_Category"],as_index = False)["Amount"].sum().sort_values(by ="Amount", ascending = False).head(15)

sns.set(rc = { 'figure.figsize' : (25,7)})
sns.barplot(data = product_sales,x= "Product_Category", y = "Amount")
```
